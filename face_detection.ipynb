```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

fd = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
img=cv2.imread("images/yaroslav_01.png")
```

```python
 plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
faces=fd.detectMultiScale(gray, 1.1, 4)
```

```python
faces
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0,255,0), 2)
```

```python
 plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
#write a face detection function

def detect_faces(img,scaleFactor,minNeighbors):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    faces=fd.detectMultiScale(gray, scaleFactor, minNeighbors)
    
    
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (0,255,0), 2)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

```

```python
img=cv2.imread("images/yaroslav_02.jpg")
detect_faces(img,2,4)
```

```python
img=cv2.imread("images/yaroslav_01.png")
detect_faces(img,2,4)
```
